```python
import pickle
import os

# Cargar el archivo grande
with open('matriz_recomendaciones_long.pkl', 'rb') as f:
    data = pickle.load(f)

# Calcular el tamaño de cada parte en bytes (20 MB)
part_size = 20 * 1024 * 1024

# Convertir el objeto a bytes para dividirlo
data_bytes = pickle.dumps(data)

# Dividir el archivo en partes más pequeñas
parts = [data_bytes[i:i + part_size] for i in range(0, len(data_bytes), part_size)]

# Guardar cada parte en un archivo separado
for i, part in enumerate(parts):
    with open(f'part_{i}.pkl', 'wb') as f:
        f.write(part)
```

```python
import pickle

# Combinar las partes
data_bytes = bytearray()
num_parts = len([name for name in os.listdir() if name.startswith('part_')])

for i in range(num_parts):
    with open(f'part_{i}.pkl', 'rb') as f:
        part = f.read()
        data_bytes.extend(part)

# Convertir los bytes combinados de nuevo al objeto original
data = pickle.loads(data_bytes)

data.head()
```
